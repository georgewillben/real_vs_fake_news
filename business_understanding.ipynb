## What is fake news?
Fake news is a buzzword that typically means false information that is circulated as if it is real news. It is especially prevelant on social media, can be generated by anyone, and can cause confusion or even become dangerous. Some examples of fake news could be edited recordings of politicians, an internet meme with false historical information, or a fake title put on an actual photo and published to a news website.

## Why would someone make fake news?
Fake news can be created to slander influential individuals, companies, or groups of people. It can also be used to spread opinions, spread conspiracy theories, and much more [2].
<br>

## Why would someone share fake news?
Fake news can be spread because of strong feelings of shock, terror, humour, hate, excitement, and a host of other emotions. It can also be used to spread opinions, such as political views. Finally it can be used to increase traffic to websites and potentially gain ad revenue or help sell products.

## Profitability of classifying fake news

Some companies such as Google have recently put in place ways to lower the impact of fake news. Facebook said it will work with third party companies to handle fake news [1].
